```python
import pandas as pd

# Load the dataset with a specific encoding
filepath = "C:/Users/thale/Desktop/TP Test 2/Informal_Settlement_Programme_-1898186875815614491 (1).csv"
data = pd.read_csv(filepath, encoding='latin1')  # or use 'ISO-8859-1'

# Display the column names
print("Column Names: ", data.columns)

# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Define the column name to sort
column_name = 'UNIREFCODE'
if column_name in data.columns:
    column_sort = data[column_name].tolist()

    def insertion_sort(array):
        for i in range(1, len(array)):
            key = array[i]
            j = i - 1
            while j >= 0 and key < array[j]:
                array[j + 1] = array[j]
                j -= 1
            array[j + 1] = key
        return array

    sorted_col = insertion_sort(column_sort)
    print("Sorted Column: ", sorted_col)
else:
    print(f"Column '{column_name}' not found in the data")
```
