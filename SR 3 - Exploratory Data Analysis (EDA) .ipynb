```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset and inspect the structure
file_path = "C:/Users/thale/Desktop/TP Test 2/allbucket1.csv"
data = pd.read_csv(file_path, encoding='latin1')  # Specifying the encoding parameter

# Display the column names and first few rows to confirm the structure
print("Column Names: ", data.columns)
print(data.head())

# Drop any rows with missing values
cleaned_data = data.dropna()

# Display the first few rows of the processed data
print(cleaned_data.head())

# Filter data for Burwan community development block
# Note: Adjusted to filter based on `admin_div` as an example, since `latitude` was used incorrectly before
burwan_data = cleaned_data[cleaned_data['admin_div'] == 'Burwan']

# Pie Chart: Distribution of Human Settlement (using `english_label` as an example)
plt.figure(figsize=(10, 6))
burwan_data['english_label'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Human Settlement in Burwan Community Development Block')
plt.ylabel('')
plt.show()

# Bar Graph: Counts of unique values in the `country` column
plt.figure(figsize=(10, 6))
burwan_data['country'].value_counts().plot(kind='bar')
plt.title('Bar Graph of Countries in Burwan Community Development Block')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Histogram: Distribution of the `latitude` column (assuming it is numeric)
plt.figure(figsize=(10, 6))
burwan_data['latitude'] = pd.to_numeric(burwan_data['latitude'], errors='coerce')
burwan_data['latitude'].dropna().plot(kind='hist', bins=20)
plt.title('Histogram of Latitude in Burwan Community Development Block')
plt.xlabel('Latitude')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot: Relationship between `latitude` and `longitude`
plt.figure(figsize=(10, 6))
burwan_data['longitude'] = pd.to_numeric(burwan_data['longitude'], errors='coerce')
plt.scatter(burwan_data['latitude'], burwan_data['longitude'])
plt.title('Scatter Plot between Latitude and Longitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()

```

```python
print(data.head())

# Drop rows with missing values
data.dropna(inplace=True)

# Select numerical and categorical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Plot distribution charts for numerical columns
plt.figure(figsize=(18, 6))
for i, column in enumerate(numeric_columns):
    plt.subplot(1, len(numeric_columns), i+1)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Plot categorical columns
num_categorical_columns = len(categorical_columns)
num_rows = (num_categorical_columns + 1) // 2  # Calculate number of rows needed
plt.figure(figsize=(18, num_rows * 6))
for i, column in enumerate(categorical_columns):
    plt.subplot(num_rows, 2, i+1)
    sns.countplot(x=column, data=data)
    plt.title(f'Counts of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
